#### 1: Concatenate the numerical and categorical dataframes

````python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.inspection import permutation_importance
import matplotlib.gridspec as gridspec
import scipy.stats as stats
````

````python
numerical = pd.read_csv('numerical.csv')
categorical = pd.read_csv('categorical.csv')
data = pd.concat([numerical, categorical], axis=1)
data.head()
````

````python
data.describe().T
````

#### 2: Split the data into training and test sets

````python
X = data.drop(['total_claim_amount'], axis=1)
y = data['total_claim_amount']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1989)
X_train.shape, X_test.shape
````

#### 3: Separate the training and test sets into numerical and categorical subsets

````python
X_train_num = X_train.select_dtypes(include=np.number)
X_test_num = X_test.select_dtypes(include=np.number)
X_train_cat = X_train.select_dtypes(exclude=np.number)
X_test_cat = X_test.select_dtypes(exclude=np.number)
X_train_num.shape, X_train_cat.shape, X_test_num.shape, X_test_cat.shape
````

#### 4: Scale the Numerical Data

````python
scaler = MinMaxScaler().fit(X_train_num)
X_train_num_scaled = scaler.transform(X_train_num)
X_test_num_scaled = scaler.transform(X_test_num)
````

````python
X_train_num_scaled_df = pd.DataFrame(X_train_num_scaled, columns = X_train_num.columns, index=X_train_num.index) 
X_test_num_scaled_df = pd.DataFrame(X_test_num_scaled, columns = X_test_num.columns, index=X_test_num.index) 
X_train_num_scaled_df
````

#### 5: Encode categorical variables - ordinal

````python
ordinal_mappings = {
    "coverage": {"Basic": 0, "Extended": 1, "Premium": 2},
    "employmentstatus": {"Unemployed": 0, "Employed": 1, "Medical Leave": 0, "Disabled": 0, "Retired": 0},
    "location_code": {"Rural": 0, "Suburban": 1, "Urban": 2},
    "vehicle_size": {"Small": 0, "Medsize": 1, "Large": 2},
    "education": {"High School or Below": 0, "College": 1, "Bachelor": 2, "Master": 3, "Doctor": 4},
    'policy_level': {'L1':1, 'L2':2, 'L3':3}
}
X_train_cat_encoded = X_train_cat.replace(ordinal_mappings)
X_test_cat_encoded = X_test_cat.replace(ordinal_mappings)
````

````python
pd.DataFrame(X_train_cat_encoded)
````

#### 6. One-hot encode categorical variables

````python
encoder = OneHotEncoder(drop='first')
X_train_cat_ohe = encoder.fit_transform(X_train_cat_encoded)
X_test_cat_ohe = encoder.transform(X_test_cat_encoded)
````

````python
X_train_cat_df = pd.DataFrame(X_train_cat_ohe.toarray(), columns=encoder.get_feature_names_out(X_train_cat_encoded.columns), index=X_train_cat_encoded.index)
X_test_cat_df = pd.DataFrame(X_test_cat_ohe.toarray(), columns=encoder.get_feature_names_out(X_test_cat_encoded.columns), index=X_test_cat_encoded.index)
X_train_cat_df
````

#### 7. Concatenate the processed numerical data and the one-hot encoded categorical data

````python
X_train_processed = pd.concat([X_train_num_scaled_df, X_train_cat_df], axis=1)
X_test_processed = pd.concat([X_test_num_scaled_df, X_test_cat_df], axis=1)
X_train_processed
````

#### 8. Explore KNeighborsRegressor

````python
def k_value_analysis(X_train_processed, y_train, y_test): 
    k_values = list(range(1, 20))
    mse_scores = []

    for k in k_values:
        knn_model = KNeighborsRegressor(n_neighbors=k)
        knn_model.fit(X_train_processed, y_train)
        knn_pred = knn_model.predict(X_test_processed)
        mse = mean_squared_error(y_test, knn_pred)
        mse_scores.append(mse)

    # Plotting the MSE scores for different k values
    plt.figure(figsize=(5, 3))
    plt.plot(k_values, mse_scores, marker='o', linestyle='-', color='b')
    plt.title('MSE for different k values')
    plt.xlabel('k')
    plt.ylabel('MSE')
    plt.grid(True)
    plt.show()
````

````python
k_value_analysis(X_train_processed, y_train, y_test)
````

The above graph depicts the Mean Squared Error (MSE) for varying values of `k` in the KNeighborsRegressor. As we can observe:

- The MSE starts at a higher value for `k=1` which is typical since a single neighbor might be sensitive to noise in the data.
- The error then decreases and appears to stabilize around `k=5` and starts to increase gradually after that.
- Having a very low `k` (like 1) means the model might be too complex (overfitting), capturing a lot of noise. On the other hand, a very high `k` can lead to an oversimplified model (underfitting). It seems that the sweet spot lies around `k=5` to `k=6` for this dataset.


````python
def distance_metrics(X_train_processed, y_train, y_test, n_neighbors):
    distance_metrics = ['euclidean', 'manhattan', 'minkowski']
    mse_scores_distance = []

    for metric in distance_metrics:
        knn_model = KNeighborsRegressor(n_neighbors=n_neighbors, metric=metric)  
        knn_model.fit(X_train_processed, y_train)
        knn_pred = knn_model.predict(X_test_processed)
        mse = mean_squared_error(y_test, knn_pred)
        mse_scores_distance.append(mse)

    # Plotting the MSE scores for different distance metrics
    plt.figure(figsize=(5, 3))
    sns.barplot(x=distance_metrics, y=mse_scores_distance, palette='viridis')
    plt.title(f'MSE for different distance metrics (with k={n_neighbors})')
    plt.xlabel('Distance Metric')
    plt.ylabel('MSE')
    plt.grid(True, axis='y')
    plt.show()
    
    return mse_scores_distance
````

````python
distance_metrics(X_train_processed, y_train, y_test, n_neighbors = 6) # Using k=6 based on the previous analysis
````

````python
def compute_permutation_importance(X_train_processed, y_train, y_test, n_neighbors, metric, threshold = 0.01):
    knn_model = KNeighborsRegressor(n_neighbors=n_neighbors, metric=metric)  # Using best parameters from earlier analysis
    knn_model.fit(X_train_processed, y_train)
    perm_importance = permutation_importance(knn_model, X_test_processed, y_test, n_repeats=20, random_state=1989)

    # Get importance scores and feature names
    sorted_idx = perm_importance.importances_mean.argsort()
    features = X_train_processed.columns

    # Plotting feature importance
    plt.figure(figsize=(10, 12))
    plt.boxplot(perm_importance.importances[sorted_idx].T, vert=False, labels=features[sorted_idx])
    plt.title('Permutation Importance for KNeighborsRegressor')
    plt.xlabel('Importance Score')
    plt.show()
    
    # Filter features based on threshold
    selected_features = []
    for idx in sorted_idx:
        if perm_importance['importances_mean'][idx] >= threshold:
            selected_features.append(idx)

    # List of selected feature indices
    print("Selected feature indices:", selected_features)
    
    return selected_features
````

````python
selected_features = compute_permutation_importance(X_train_processed, y_train, y_test, 6, 'manhattan')
````

````python
selected_features
````

````python
def weight_metrics(X_train_processed, y_train, y_test, n_neighbors, metric):
    weights_options = ['uniform', 'distance']
    mse_scores_weights = []

    for weight in weights_options:
        knn_model = KNeighborsRegressor(n_neighbors=n_neighbors, metric=metric, weights=weight)
        knn_model.fit(X_train_processed, y_train)
        knn_pred = knn_model.predict(X_test_processed)
        mse = mean_squared_error(y_test, knn_pred)
        mse_scores_weights.append(mse)

    plt.figure(figsize=(5, 3))
    sns.barplot(x=weights_options, y=mse_scores_weights, palette='viridis')
    plt.title(f'MSE for different weight options (with k={n_neighbors} and metric={metric})')
    plt.xlabel('Weights Option')
    plt.ylabel('MSE')
    plt.grid(True, axis='y')
    plt.show()
    
    return mse_scores_weights
````

````python
weight_metrics(X_train_processed, y_train, y_test, 6, 'manhattan')
````

````python
from sklearn.model_selection import GridSearchCV #This is a utility that performs grid search combined with cross-validation.

def find_best_hyperparameters(X_train_processed, y_train, y_test):
    # Define the hyperparameters and their possible values
    param_grid = {'n_neighbors': list(range(1, 50)),
                  'weights': ['uniform', 'distance'],
                  'metric': ['euclidean', 'manhattan', 'minkowski']}

    # Initialize the KNeighborsRegressor
    knn_regressor = KNeighborsRegressor()

    # Set up GridSearchCV
    grid_search = GridSearchCV(knn_regressor, param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1, verbose=1)

    # Fit the model with the parameter candidates
    grid_search.fit(X_train_processed, y_train)

    # Get the best parameters and best score
    best_params = grid_search.best_params_
    best_score = grid_search.best_score_
    
    return best_params, -best_score  # Negating the score to get positive MSE
````

````python
# Here, we're initializing the GridSearchCV:

# knn_regressor: The model we want to tune.
# param_grid: The hyperparameters and their possible values.
# cv=5: We're performing 5-fold cross-validation.
# scoring='neg_mean_squared_error': The performance metric. It's negative because GridSearchCV tries to maximize the score, so by convention, loss functions are negative.
# n_jobs=-1: This means use all available processors to speed up the computation.
# verbose=1: Display messages while running.
````

````python
best_params, best_score = find_best_hyperparameters(X_train_processed, y_train, y_test)
print('best_params: ') 
print(best_params)
print('best_score: ')
print(best_score)
````

### Observations:

The Grid Search Cross-Validation determined that the best hyperparameters for the KNeighborsRegressor on the given dataset are:

- Metric: manhattan
- Number of Neighbors (`n_neighbors`): 12
- Weights: distance

The best Mean Squared Error (MSE) achieved with these hyperparameters is approximately 0.36.

By using Grid Search, we've systematically determined the optimal hyperparameters without the need for manual inspection of graphs.


**Observations Comparison:**

**Visual Inspection:**
Based on our earlier visual inspection of the MSE graph against varying values of `n_neighbors`, we observed that the model seemed to stabilize and perform well around `n_neighbors = 5` to `n_neighbors = 6`. The distance metric `manhattan` seemed to slightly outperform the others, and the `uniform` weighting was initially chosen.

**Grid Search Cross-Validation:**
Upon performing a systematic search using GridSearchCV, the best hyperparameters determined were `n_neighbors = 12`, `metric = manhattan`, and `weights = distance`. This suggests that giving more weight to closer neighbors (distance weighting) and considering a larger neighborhood (with 20 neighbors) yielded the lowest mean squared error in cross-validation.

**Comparison:**
It's interesting to note the divergence in the optimal `n_neighbors` value between the visual inspection and the grid search. One reason could be that the visual inspection was based on a single train-test split, while GridSearchCV used 5-fold cross-validation, leading to a more robust estimation of performance. Additionally, the grid search took into account combinations of all hyperparameters, which might have led to the identification of a more optimal set of parameters.

````python
def train_and_test_models(models, X_train, y_train, X_test, y_test):
    results = {}
    
    for model_name, model in models.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

        r2 = r2_score(y_test, y_pred)
        mse = mean_squared_error(y_test, y_pred)
        
        results[model_name] = (r2, mse, y_pred)
        
    return results
````

````python
from sklearn.linear_model import Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

models_to_evaluate = {
    "Linear Regression": LinearRegression(),
    "KNeighbors Regressor": KNeighborsRegressor(**best_params),
    "MLP Regressor": MLPRegressor(max_iter=500, random_state=1989),
    "Ridge Regression": Ridge(),
    "Decision Tree Regressor": DecisionTreeRegressor(random_state=1989),
    "Random Forest Regressor": RandomForestRegressor(random_state=1989),
    "Gradient Boosting Regressor": GradientBoostingRegressor(random_state=1989)
}
````

````python
model_results = train_and_test_models(models_to_evaluate, X_train_processed, y_train, X_test_processed, y_test)
pd.DataFrame(model_results).T
````

````python
def plot_model_diagnostics(results, y_test):
    for model_name, (r2, mse, y_pred) in results.items():
        residuals = y_test - y_pred
        
        # Setting up the subplot grid
        fig = plt.figure(figsize=(20, 4))
        spec = gridspec.GridSpec(ncols=4, nrows=1, figure=fig)
        
        # Residuals vs Fitted
        ax1 = fig.add_subplot(spec[0, 0])
        ax1.scatter(y_pred, residuals, alpha=0.5)
        ax1.axhline(y=0, color='r', linestyle='--')
        ax1.set_title(f'Residuals vs Fitted for {model_name}')
        ax1.set_xlabel('Fitted values')
        ax1.set_ylabel('Residuals')

        # Histogram of Residuals
        ax2 = fig.add_subplot(spec[0, 1])
        ax2.hist(residuals, bins=30, alpha=0.8, color='skyblue', edgecolor='black')
        ax2.set_title(f'Histogram of Residuals for {model_name}')
        ax2.set_xlabel('Residuals')
        ax2.set_ylabel('Frequency')

        # QQ-Plot
        ax3 = fig.add_subplot(spec[0, 2])
        stats.probplot(residuals, dist="norm", plot=ax3)
        ax3.set_title(f'QQ-Plot for {model_name}')

        # Actual vs Predicted
        ax4 = fig.add_subplot(spec[0, 3])
        ax4.scatter(y_test, y_pred, alpha=0.5)
        ax4.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
        ax4.set_title(f'Actual vs Predicted for {model_name}')
        ax4.set_xlabel('Actual values')
        ax4.set_ylabel('Predicted values')

        # Adjust layout
        plt.tight_layout()
        plt.show()
````

````python
 plot_model_diagnostics(model_results, y_test)
````

````python
def plot_comparison_metrics(results):
    model_names = list(results.keys())
    r2_scores = [results[model_name][0] for model_name in model_names]
    mse_scores = [results[model_name][1] for model_name in model_names]
    
    # Setting up the subplot grid
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))
    
    # Define a color palette
    colors = plt.cm.tab10(np.linspace(0, 1, len(model_names)))
    
    # Plotting R2 scores
    ax1.bar(model_names, r2_scores, color=colors)
    ax1.set_title('R2 Scores Comparison')
    ax1.set_ylabel('R2 Score')
    ax1.set_ylim(0, 1)  # Assuming R2 is between 0 and 1
    ax1.grid(True, axis='y')
    
    # Plotting MSE scores
    ax2.bar(model_names, mse_scores, color=colors)
    ax2.set_title('MSE Scores Comparison')
    ax2.set_ylabel('Mean Squared Error (MSE)')
    ax2.set_xlabel('Models')
    ax2.grid(True, axis='y')
    
    # Adjust layout
    plt.tight_layout()
    plt.xticks(rotation=45)
    plt.show()
````

````python
plot_comparison_metrics(model_results)
````
